PROJECT 2: IMAGE CLASSIFICATION WITH CNN (MNIST)

```python
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.model_selection import train_test_split

# Load MNIST-like digits dataset (8x8 images)
digits = load_digits()
X_img = digits.data
y_img = digits.target

# Normalize
scaler = StandardScaler()
X_img_scaled = scaler.fit_transform(X_img)

# Split data
X_train_img, X_test_img, y_train_img, y_test_img = train_test_split(
    X_img_scaled, y_img, test_size=0.2, random_state=42
)

# Train neural network
nn_model = MLPClassifier(
    hidden_layer_sizes=(128, 64),
    activation='relu',
    max_iter=50,
    random_state=42
)
nn_model.fit(X_train_img, y_train_img)

# Predictions
y_pred_img = nn_model.predict(X_test_img)

# Evaluation
print(f"\nAccuracy: {accuracy_score(y_test_img, y_pred_img):.4f}")
print("\nClassification Report:")
print(classification_report(y_test_img, y_pred_img))
```
